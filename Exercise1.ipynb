Question10.
Develop a menu ordering program for Yummy Restaurant. Show the following menu to the user and
and let the user select one. If the given input alphabet is not in the menu, print 'enter the menu
again:' and receive another input.

```python

print("-Bruger(enter b)")
print("-Chicken(enter c)")
print("-Pizza(enter p)")
choice = input("Choose a menu(enter b,c,p) : ")

while(choice not in 'bcp'):
    choice = input("Choose a menu(enter b,c,p) : ")

if(choice == 'b'):
    print("You choose burger")
elif(choice=='c'):
    print("You choose chicken")
else:
    print("You choose pizza")
```

Question11.
print the biggest number between three

```python
numbers = input("Enter there number: ")
nums = numbers.split()
num1,num2,num3 = nums

num1 = int(num1)
num2 = int(num2)
num3 = int(num3)

if(num1>num2 and num1>num3):
    print(num1,"is greatest")
elif(num2>num1 and num2>num3):
    print(num2,"is greatest")
else:
    print(num3,"is greatest")
```
